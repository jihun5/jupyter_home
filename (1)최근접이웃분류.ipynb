산점도 그래프

파이썬에서 그래프를 그리는 대표적인 패키지는 맷플롯립,

import는 따로 만들어둔 파이썬 패키지를 사용하기 위해 불러 오는 명령

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel("lenght") # x축
plt.ylabel("weight") # y축
plt.show()
```

그래프가 일직선상으로 퍼져 있는 상태를 선형적으로 부른다.

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
# zip함수는 나열된 리스트에서 각각 하나씩 원소를 꺼내 반환한다.
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
print(fish_data)
```

2차원 리스트 형성

```python
fish_target = [1] * 35 +[0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

k-군집분류 사용



```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

```python
kn.predict([[30, 600]])
```

```python
print(kn._fit_X)
```

```python
print(kn._y)
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
print(35/49)
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  # k - 최근접 이웃 개수 설정
  kn = KNeighborsClassifier(n_neighbors=n)
  kn.fit(fish_data, fish_target)
  # 점수 계산
  score = kn.score(fish_data, fish_target)
  # 100%의 정확도에 미치지 못하는 이웃 개수 출력
  if score < 1:
    print(n, score)
    break
```
